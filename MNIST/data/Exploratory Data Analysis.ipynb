```python
import pandas as pd
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Read and Explore Labels Data

```python
data_path = 'extracted_data/'
```

### Read Labels in the Training Set and Plot the Histogram

```python
# Read Training Set
data_train = pd.read_csv(data_path + 'train-labels.csv',header =None)
paths_train = data_train[0]
labels_train = data_train[1]
```

```python
# Visualize the Distribution
plt.hist(labels_train)
plt.title("Distribution of images by labels - Training set")
plt.xlabel('Digit Labels')
plt.ylabel('Counts')
plt.show()
```

### Read Labels in the Test Set and Plot the Histogram

```python
# Read Testing Set
data_test = pd.read_csv(data_path + 'test-labels.csv',header =None)
paths_test = data_test[0]
labels_test = data_test[1]
```

```python
# Visualize the Distribution
plt.hist(labels_test)
plt.title("Distribution of images by labels - Test set")
plt.xlabel('Digit Labels')
plt.ylabel('Counts')
plt.show()
```

## Read Sample Image Data

Here we read only 10 images 

```python
imgs = []
dim = (32,32)
for i in range(0,10):
    imfile = data_path + paths_test[i]
    im = cv2.imread(imfile)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    resized = cv2.resize(im,dim,interpolation = cv2.INTER_AREA) # Resize the image
    imgs.append(resized)
imgs = np.asarray(imgs)
```

```python
imgs.shape
```

### Plot images

```python
def plot_images(images, labels):
    assert len(images) == len(labels) == 9
    
    # Create figure with 3x3 sub-plots.
    fig, axes = plt.subplots(3, 3)
    fig.subplots_adjust(hspace=0.3, wspace=0.3)

    for i, ax in enumerate(axes.flat):
        # Plot image.
        ax.imshow(images[i],cmap = 'gray')

        # Show labels
        xlabel = labels[i]

        # Show the classes as the label on the x-axis.
        ax.set_xlabel(xlabel)
        
        # Remove ticks from the plot.
        ax.set_xticks([])
        ax.set_yticks([])
    
    # Ensure the plot is shown correctly with multiple plots
    # in a single Notebook cell.
    plt.show()
```

```python
plot_images(imgs[0:9], labels_test[0:9])
```
