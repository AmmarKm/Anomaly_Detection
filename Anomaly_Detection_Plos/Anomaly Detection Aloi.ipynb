```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('aloi-unsupervised-ad.csv',delimiter=',')
data_height, data_width = data.shape
```

```python
# Read features and labels
features = data[:,:data_width-1]
labels = data[:,data_width-1]
features.shape, labels.shape
```

```python
feature_hist(features, data_width - 1)
```

```python
def feature_hist(data, num_features, cols= 4, num_bins = 40):
    """
    Plot histograms of each feature
    """
    rows = round(num_features/4)
    fig, axes = plt.subplots(rows, cols,figsize=(15,15))
    fig.subplots_adjust(hspace=0.3, wspace=0.01)

    for i, ax in enumerate(axes.flat): 
        if i < num_features:
            ax.hist(data[:,i],num_bins)
            xlabel = "Feature: {0}".format(i+1)
            # Show the classes as the label on the x-axis.
            ax.set_xlabel(xlabel)
            ax.set_xlim(min(data[:,i]),max(data[:,i]))

            # Remove ticks from the plot.
            # ax.set_xticks([])
        ax.set_yticks([])

    plt.show()
```

```python
### Mean Shift and Normalization
means = np.mean(features,axis = 0) # Take the mean of each column
maxs = np.mean(features, axis = 0)
mins = np.mean(features, axis = 0)
ranges = maxs - mins # Get the range of each column
ranges[ranges == 0] = 1 # We cannot have 0 in the denominator

# Mean Shift
features_shifted = features - means
# Normalize
features_normalized = np.divide(features_shifted,ranges)
```

```python
feature_hist(features_normalized, data_width - 1)
```
